```python
import os
import docx  # To read DOCX files
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
directory_path = ('/content/drive/MyDrive/Document')
!ls {directory_path}
```

```python
# Step 1: Function to read DOCX files from Google Drive
def read_docx_file(file_path):
    doc = docx.Document(file_path)
    text = "\n".join([para.text for para in doc.paragraphs])
    return text

def read_documents_from_directory(directory_path):
    documents = []
    filenames = []
    for filename in os.listdir(directory_path ):
        if filename.endswith('.docx'):
            file_content = read_docx_file(os.path.join(directory_path, filename))
            documents.append(file_content)
            filenames.append(filename)
    return documents, filenames

```

```python
# Step 2: Convert documents into TF-IDF vectors
def compute_tfidf_vectors(documents):
    vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = vectorizer.fit_transform(documents)
    return tfidf_matrix
```

```python
# Step 3: Compute Cosine Similarity
def compute_cosine_similarity(tfidf_matrix):
    return cosine_similarity(tfidf_matrix)
```

```python
# Step 4: Visualize the similarity heatmap
def plot_similarity_heatmap(cosine_sim_matrix, filenames):
    df = pd.DataFrame(cosine_sim_matrix, index=filenames, columns=filenames)
    plt.figure(figsize=(10, 8))
    sns.heatmap(df, annot=True, fmt='.2f', cmap='coolwarm')
    plt.title('Cosine Similarity Between Documents')
    plt.show()
```

```python
# Main Execution
if __name__ == "__main__":

    # Step 1: Load the DOCX files from the folder
    documents, filenames = read_documents_from_directory(directory_path)

    # Step 2: Compute TF-IDF vectors
    tfidf_matrix = compute_tfidf_vectors(documents)

    # Step 3: Compute cosine similarity
    cosine_sim_matrix = compute_cosine_similarity(tfidf_matrix)

    # Step 4: Visualize the similarity heatmap
    plot_similarity_heatmap(cosine_sim_matrix, filenames)

    # Optional: Print similarity matrix as DataFrame
    similarity_df = pd.DataFrame(cosine_sim_matrix, index=filenames, columns=filenames)
    print(similarity_df)
```
